## Introduction

Here I give a motivation for the code I have written and some of the general principles used. The technical and mathematical details can be found in an Appendix at the bottom (at some point). Simplier examples using explicit, implicit, and Monte Carlo solvers for the simplier diffusion equation can be found in the notebook Diffusion Examples and corresponding julia file diffusion_solvers.jl. 

Cosmic rays and other energetic particles of density $u(x,t,E)$ diffuse in the galaxy with an energy dependent diffusion constant $D(E) \propto E^{-s}$. As they do so they also lose energy at a rate $\frac{dE}{dt} = - b(E)$. One way to model this phenomina is by solving a diffusion equation with an advective term to describe energy loss: 

$$\frac{\partial u}{\partial t} = D(E)\frac{\partial^2 u}{\partial x^2} + \frac{\partial}{\partial E} \left[b(E) u\right]$$

There are many techniques to solve partial differential equations with > 2 variables, but they can roughly be categorized by explicit methods (the solution at the next point in time depends only on the information at the last point in time), implicit methods (the solution at the next time step is a linear combination of the currently unknown solution at other points), and Monte Carlo methods which utilize a random walk. Here I have attempted to impliment an explicit finite difference and Monte Carlo solver where the latter is implimented to solve an intrinsically implicit finite difference scheme. 

## Solvers

Examples to all the solvers referenced here are shown below. 

To take advantage of the finite difference solvers a rectangular numerical domain is generated in position and energy for the solution $u_{i,j,k}$ where where $(i,j,k) \rightarrow $(space, time, energy) respectively and $1 < i < nx$, $1 < j < nt$, $1 < k < nE$. Given these definitions we can solve the problem given some initial information $u_{i,1,k}$ and boundary conditions $u_{1,j,k}, u_{nx,j,k}, u_{i,j,1}, \textrm{ and }\ u_{i,j,nE}$ and an algorithm to propogate information from $j = 1$ to $j = nt$. 

#### Explicit Solver

The explicit solver uses: 
$$ u_{i,j+1,k} = \alpha_k\left(u_{i+1,j,k} + u_{i-1,j,k}\right) + (1 - 2\alpha_k)u_{i,j,k} + \beta_{k+1}u_{i,j,k+1} - \beta_{k-1}u_{i,j,k-1}$$

where $\alpha_k = dt\ D_k/dx^2$ and $\beta_k = dt\ b_k/2 dE$ and $D_k \mathrm{ and } b_k$ are $D(E)$ and $b(E)$ evaluated at energy point $k$ respectively. Unfortunately, stability analysis shows that this method is stable only for specific grid configurations which satisfy the equation $\left| w \right|^2 < 1$ where: 
$$ \left| w_k \right|^2 = \left(1 - 2\alpha_k\left(1 - cos(q_x dx\right)\right)^2 + 4\beta_k^2sin^2(q_E dE)$$
for $cos(q_x dx) \simeq 1$ which correponds to the Forier components of the solution which have a wavelength $\lambda_x \simeq 4\ dx$, the solution will become unstable. 

The explicit solver is implimented in the file advection_diffusion_solver.jl with the function exp_advection_diffusion_solver.

#### Monte Carlo Solver

To impliment the Monte Carlo solver we need a slightly difference finite difference scheme because we require that all of the coefficients of $u_{i+1,j,k}$ etc are positive. Here we impliment a scheme which uses the trapezoidal integration method. If we define: 
$$\frac{\partial u(x,t,E)}{\partial t} = F(x,t,u(x,t,E))$$

then 
$$u(x,t+dt,E) \simeq u(x,t,E)+\frac{dt}{2} \left(F(x,t+dt,u(x,t+dt,E)) + F(x,t,u(x,t,E))\right)$$
implimenting this approximation for the advection diffusion equation we get the following finite difference scheme. 
$$u_{i,j+1,k} = \frac{\alpha_k}{2(1+\alpha_k)}\left[u_{i+1,j+1,k} + u_{i+1,j,k} + u_{i-1,j+1,k} + u_{i-1,j,k} \right] + 
\frac{\frac{1}{2}-\frac{\alpha_k}{2}+\beta_{k+1}}{2(1+\alpha_k)}\left[u_{i,j+1,k+1} + u_{i,j,k+1}\right] + \frac{\frac{1}{2}-\frac{\alpha_k}{2}-\beta_{k-1}}{2(1+\alpha_k)}\left[u_{i,j+1,k-1} + u_{i,j,k-1}\right]$$

where we have made the approximation that $u_{i,j,k} \simeq \frac{1}{4}\left[u_{i,j+1,k+1} + u_{i,j,k+1} + u_{i,j+1,k-1} + u_{i,j,k-1}\right]$. Now notice that all of the coefficients for $u_{i,j+1,k+1}$ etc. are positive as long as $\frac{1}{2} - \frac{\alpha_k}{2} - \beta_{k-1} > 0$. We can interpret these coefficients as the probability that a random walker position at the point $(i,j+1,k)$ will move to any of the neighboring points specified in the equation given proper normalizations. Notice there are no limits to the grid spacing as was required by the explicit solver. This allows us to solve the PDE using less spacial and energy points. 

The Monte Calro solver is implimented in the file advection_diffusion_solver.jl with the function MC_advection_diffusion_solver. 

### Problem Setup

As stated above, one requires a finite difference algorithm as well as initial and boundary conditions. Here we assume the solution is zero along all boundaries, and we will make our computational domain large enough that the solution should not interact with the boundaries to eliminate edge effects. To specifiy an initial condition, I assume that there is an inital population of particles that are distributed in the middle of the spatial domain with a Gaussian distribution. These particles have a power law $u(E) \propto E^{-2}$ spectrum with specified spectral cutoff energies. The initial spacial function and energy spectrum are found in advection_diffusion_solvers (the functions u0 and energy_spectrum) and are plotted below. The energy dependence and energy loss rates are also specified there (the functions calc_diffusion_constant and calc_energy_loss_rate). As stated above $D(E) \propto E^{-s}$ where in general $0.3 \lesssim s \lesssim 0.7$ and I have chosen $s = 0.4$. The energy loss rate was chosen to be $b(E) \propto \sqrt{E}$ arbitrarilly.

```julia
using PyPlot
```

```julia
include("advection_diffusion_solvers.jl");
```

```julia
setup_xi = 0.0; 
setup_xf = 100.0; 
setup_nx = 1000; 

setup_Ei = 0.0; 
setup_Ef = 50.0;
setup_nE = 500; 

setup_x = linspace(setup_xi,setup_xf,setup_nx); 
setup_E = linspace(setup_Ei,setup_Ef,setup_nE); 
```

```julia
plot(setup_x,map(a-> u0(a,0.5*(setup_xi+setup_xf)),setup_x));
title("Initial Spatial Distribution"); 
ylabel("u(x,0,15)");
xlabel("x");
```

```julia
plot(setup_E,map(b->energy_spectrum(b,15.,40.),setup_E));
title("Initial Power Law Energy Spectrum");
ylabel("u(50,0,E)");
xlabel("E");
```

Note that there are energy cutoffs to prevent divergences in the spectrum and interaction with the boundaires

```julia
plot(setup_E,map(calc_diffusion_constant,setup_E))
axis([setup_Ei,setup_Ef,0,6])
title("Diffusion Constant")
ylabel("D(E)");
xlabel("E");
```

```julia
plot(setup_E,map(calc_energy_loss_rate,setup_E));
title("Energy Loss Rate"); 
ylabel("b(E)"); 
xlabel("E");
```

## Examples 

Here I've implimented both the solvers to compare the results. The initial condition and results are displayed

```julia
#Monte Carlo Solver 
mc_xi = 40.0; 
mc_xf = 60.0; 
mc_nx = 100; 

Ei = 0.0; 
Ef = 20.0;

mc_nE = 70;

spec_E_min = 5.0;
spec_E_max = 17.;

tf = 0.1;
```

```julia
mc_sol, mc_x, mc_E, mc_ic = MC_advection_diffusion_solver(mc_xi,mc_xf,mc_nx,Ei,Ef,mc_nE,spec_E_min,spec_E_max,tf,1000);
```

```julia
#explicit solver 
e_xi = 0.0;
e_xf = 100.0;
e_nx = 500; 

e_nE = 500;

e_sol, e_x, e_E, e_ic = exp_advection_diffusion_solver(e_xi,e_xf,e_nx,Ei,Ef,e_nE,spec_E_min,spec_E_max,tf,100); 
```

```julia
#Monte Carlo initial condition
imshow(mc_ic,extent=[Ei,Ef,mc_xi,mc_xf]);
```

```julia
#explicit solver initial condition 
min_x_index = get_x_index(e_xi,e_xf,mc_xi,e_nx); 
max_x_index = get_x_index(e_xi,e_xf,mc_xf,e_nx);
e_x_range = min_x_index:max_x_index;
imshow(e_ic[e_x_range,:],extent = [Ei,Ef,mc_xi,mc_xf]);
colorbar();
```

```julia
#Monte Carlo solution 
imshow(mc_sol,extent=[Ei,Ef,mc_xi,mc_xf]);
colorbar(); 
```

```julia
#explicit solver solution 
min_x_index = get_x_index(e_xi,e_xf,mc_xi,e_nx); 
max_x_index = get_x_index(e_xi,e_xf,mc_xf,e_nx);
e_x_range = min_x_index:max_x_index;
imshow(e_sol[e_x_range,:],extent = [Ei,Ef,mc_xi,mc_xf]);
colorbar();
```

```julia
#difference between initial condition and Monte Carlo solution
imshow(mc_ic - mc_sol,extent=[Ei,Ef,mc_xi,mc_xf])
colorbar();
```

```julia
#difference between initial condition and explicit solution
min_x_index = get_x_index(e_xi,e_xf,mc_xi,e_nx); 
max_x_index = get_x_index(e_xi,e_xf,mc_xf,e_nx);
e_x_range = min_x_index:max_x_index;
imshow(e_ic[e_x_range,:]-e_sol[e_x_range,:],extent = [Ei,Ef,mc_xi,mc_xf]);
colorbar();
```
