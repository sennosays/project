## Introduction

Here I've provided some simple examples of the different types of methods used to solve PDE's including an explict solver, two implicit solver, and a Monte Carlo solver which utilizes the finite difference technique used for one of the implicit solvers. This notebook eliminates many of the more complicated features of the advection diffusion solver to differentiate the different solving methods. We are trying to solve the simplier equation: 
$$ \frac{\partial u}{\partial t} = D \frac{\partial^2 u}{\partial x^2}$$ 
Where D is now a constant

## Solvers

Each solving method has its own benifits and drawbacks. The explicit solver is the most straightforward to impliment, but can be sensitive to the construction of the numerical domain (choosing a proper dx and dt) to remain stable. An implicit solver does not suffer from such instabilities so that any reasonable dx and dt can be used, implicit methods require the solution of a large system of linear equations oftentimes in the form of matrix inversion. This can be expensive in both time and memory. Monte Carlo methods can be implimented to solve either an explicit or implicit finite difference algorithm so suffers from the same stability criteria, but also requires a large amount of sample to converse to the same result as an explicit or implicit solver. However, for well posed problems the amount of trials may be relatively small or can be distributed with very little overhead. 

All of these solvers take different approches to solving the equation: 
$$ \frac{\partial u(x,t)}{\partial t} = F(x,t,u)$$

and note that we have a rectangular numerical domain where the index $i$ locates a point in space and the index $j$ locates a point in time for $nx$ spatial points and $nt$ temporal points. The spacings $dt$ and $dx$ are calculated from $dx = (x_f - x_i)/(nx+1)$ and $dt = t_f/nt$

#### Explicit Solver

Approximation: $ u(x,t+dt) \simeq u(x,t) + dt\ F(x,t,u(x,t))$

Finite Difference Equation: $ u_{i,j+1} = \lambda u_{i+1,j} + (1-2\lambda)u_{i,j} + \lambda u_{i-1,j}$
where $\lambda \equiv \frac{dt D}{dx^2}$

For stability we require that $\lambda < 0.5$

#### Implicit Solver

Approximation: $u(x,t+dt) \simeq u(x,t) + dt\ F(x,t+dt,u(x,t+dt))$

Finite Difference Equation: $-\lambda u_{i+1,j+1} + (1+2\lambda)u_{i,j+1} - \lambda u_{i-1,j+1} = u_{i,j}$

No stability requirement 

#### Crank -Nicolson (Implicit Sover #2)

Approximation: $u(x,t+dt) \simeq u(x,t) + \frac{dt}{2}\left[F(x,t+dt,(u(x,t+dt)) + F(x,t,u(x,t))\right]$

Finite Difference Equation: $\lambda u_{i+1,j+1} - 2(1+\lambda)u_{i,j+1} + \lambda u_{i-1,j+1} = -\lambda u_{i+1,j} - 2(1-\lambda)u_{i,j} - \lambda u_{i-1,j}$

No stability requirement, more accurate

#### Monte Carlo Using Crank-Nicolson 

All of the above methods have required a solution to be propagated forward in time (the solution at time $j+1$ is a linear combination of the solution at $j+1$ and $j$) but the Monte Carlo method uses a different idea. Instead, a random walker starts its walk at the furthest time point $(i,j = nt)$ and can step to any of the neighboring grid points ($(i+1,j+1),(i-1,j+1),(i+1,j),(i-1,j),$ or $(i,j)$ based on a certain probability. It can be shown that the probability of the random walker stepping to a neighbroing grid point is the (normalized) coefficient of that neighbor points term in the finite difference approximation. For example, after some manipulation the the Crank-Nicolson finite difference equation becomes:
$$ u_{i,j+1} = \frac{\lambda}{2(1+\lambda)} \left[u_{i+1,j+1} + u_{i-1,j+1} + u_{i+1,j} + u_{i-1,j}\right] + \frac{1-\lambda}{1+\lambda}u_{i,j}$$

note that $4\times\frac{\lambda}{2(1+\lambda)} + \frac{1-\lambda}{1+\lambda} = 1$ so each coefficient is properly normalized. Therefore, if we have a random walker at point (5,8) the probability that the walker will next occupy the point (5,7) is $\frac{1-\lambda}{1+\lambda}$ while the probabiliy it will occupy (4,8), (6,8), (4,7), or (6,7) are is $\frac{\lambda}{2(1+\lambda)}$ each. We let this random walker move from grid point to grid point until it reaches a boundary condition $(i = 1$ or $nx)$ or the initial condition $(j = 1)$ and the number of times a walker has reached that point is increased by 1. Because we known the value of the solution at the boundaries and the initial condition after many random walks we compute the solution at (i,nt): 

$$u_{i,nt} = \sum_{i',j'} P(i',j')u_{i',j'}$$

Where $i',j'$ correspond to grid points on the boundary (the boundary includes the initial condition), $P(i',j')$ is the probability a walker starting at $(i,nt)$ will stop its walk at $(i',j')$, and $u_{i',j'}$ is the known solution located at grid point $(i',j')$
