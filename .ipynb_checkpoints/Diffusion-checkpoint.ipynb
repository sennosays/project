```julia
function ui(x::Float64)
    return sin(2pi*x)
end


function exact_solution(x::Float64,t::Float64)
    return exp(-4*pi^2*t)*sin(2*pi*x);
end
```

```julia
function explicit_solver(g,xi::Float64,xf::Float64,tf::Float64,D::Float64,N::Int = 50) 
    dx = (xf-xi)/(N+1); 
    M = 2*N; 
    dt = tf/M; 
    lambda = dt*D/dx/dx; 
    while lambda > 0.5
        M = M + min(M,200); 
        dt = tf/M; 
        lambda = dt*D/dx/dx; 
    end
    println("lambda for the explicit solver was deterined to be: ",lambda); 
    
    A = zeros(Float64,N,N); 
    for i in 1:N 
        A[i,i] = 1.0-2.0*lambda;
        if i > 1
            A[i,i-1] = lambda;
        end
        if i < N
            A[i,i+1] = lambda; 
        end
    end
    
    newu = Array(Float64,N); 
    oldu = Array(Float64,N); 
    
    x = linspace(xi+dx,xf-dx,N); 
    
    oldu = map(g,x); 
    
    for j in 1:M-1
        newu = A*oldu; 
        oldu = newu;
    end    
    
    #add boundary conditions 
    append!(oldu,[0]); 
    prepend!(oldu,[0]);
    
    return oldu; 
         
end
```

```julia
function implicit_solver(g,xi::Float64,xf::Float64,tf::Float64,D::Float64,N::Int = 50,M::Int = 50)
    dx = (xf-xi)/(N+1); 
    dt = tf/M; 
    
    lambda = dt*D/dx/dx; 
    println("lambda for the implicit solver was determined to be: ",lambda); 
    
    B = zeros(Float64,N,N); 
    for i in 1:N 
        B[i,i] = 1.0+2.0*lambda;
        if i > 1
            B[i,i-1] = -lambda;
        end
        if i < N
            B[i,i+1] = -lambda; 
        end
    end
    
    invtB = inv(B);
    
    newu = Array(Float64,N); 
    oldu = Array(Float64,N); 
    
    x = linspace(xi+dx,xf-dx,N); 
    
    oldu = map(g,x); 
    
    for j in 1:M-1
        newu = invtB*oldu; 
        oldu = newu;
    end    
    
    #add boundary conditions 
    append!(oldu,[0]); 
    prepend!(oldu,[0]);
    
    return oldu; 
end
```

```julia
D = 1.0; 
Q0 = 1.0; 

xi = -10.0; 
xf = 10.0; 
xg = 0.5; 

N = 2000;  
dx = (xf-xi)/(N+1); 
x = linspace(xi,xf,N+2); 

tf = 1.0; 
M = iceil(2.1*tf*D/dx^2)
dt = tf/M

lambda = dt*D/dx^2
temp = 1.0-2.0*lambda; 
diag = temp*ones(N+2);
off_diag = lambda*ones(N+1); 

A = spdiagm((off_diag,diag,off_diag),[-1,0,1]); 
A[1,1] = (1.0+lambda); 
A[1,2] = -5.0*lambda; 
A[1,3] = 4.0*lambda; 
A[1,4] = -lambda; 

A[N+2,N+2] = 1.0+lambda; 
A[N+2,N+1] = -5.0*lambda; 
A[N+2,N] = 4.0*lambda; 
A[N+2,N-1] = -lambda; 

new_u = Array(Float64,N+2); 
old_u = Array(Float64,N+2); 
Q = Array(Float64,N+2); 

#initialize function 
fill!(old_u,0.0);
fill!(Q,0.0); 

Q[iceil(end/2)] = Q0/dx; 


for j = 1:M
    new_u = A*old_u+dt*Q; 
    old_u = new_u; 
end
```

```julia
#interaction length 
lambda = 1.0
num_walkers = 100000; 
walker_pos = Array(Float64,num_walkers); 
num_steps = 100; 
for j in 1:num_walkers
    pos = 0.0; 
    l = -lambda*log(rand(num_steps)); 
    for i in 1:num_steps
        pos += (rand() < 0.5) ? l[i] : -l[i]; 
    end
    walker_pos[j] = pos;
end
```

```julia
maximum(walker_pos)
```

```julia
num_bins = iceil((maximum(walker_pos)-minimum(walker_pos))/lambda); 
plt.hist(walker_pos,num_bins,normed=1);
```

```julia
function random_walker(this_num_steps::Int,this_lambda::Float64)
    this_l = -this_lambda*rand(this_num_steps); 
    #steps = [(rand() < 0.5) ? this_l[i] : -this_l[i] for i in 1:this_num_steps]; 
    this_pos = 0.0; 
    for i in 1:this_num_steps
        this_pos += (rand() < 0.5) ? this_l[i] : -this_l[i]
    end
    return this_pos; 
    #return sum(steps); 
end
    
```
