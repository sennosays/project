```julia
function random_walker(this_num_steps::Int,this_lambda::Float64)
    this_l = -this_lambda*log(rand(this_num_steps)); 
    steps = [(rand() < 0.5) ? this_l[i]: -this_l[i] for i in 1:this_num_steps]; 
    
    return sum(steps); 
end

function get_prob_random_walker(dist,lambda::Float64,num_steps::Int,num_walkers::Int = 100000)
    @assert(lambda > 0.0); 
    @assert(num_steps > 2); 
    walker_pos = Array(Float64,num_walkers);  
    for j in 1:num_walkers
        walker_pos[j] = random_walker(num_steps,lambda);
    end

    new_dist, prob = hist(walker_pos,dist)
    @assert(new_dist == dist);
    prob /= num_walkers; 
    
    return prob, walker_pos; 
end

function add_prob_random_walker(dist,old_pos::Array{Float64,1},lambda::Float64,new_num_steps::Int)
    num_walkers = length(old_pos); 
    @assert(num_walkers > 1); 
    for j in 1:num_walkers
        old_pos[j] += random_walker(new_num_steps,lambda); 
    end
    new_dist , new_prob = hist(old_pos,dist);
    @assert(new_dist == dist); 
    new_prob /= num_walkers; 
    
    return new_prob; 
end
    
```

```julia
#info from the finite difference solver
tf = 1.0; 
dt = 0.0007587; 
M = iround(tf/dt);
Q0 = 1.0; 
D = 1.0; 
c = sqrt(5*2*D/dt); 
```

```julia
lambda = 2.0*D/c; 
dx = 1.5*lambda; 
x_ends = -20.:dx:20.
x = -20.+dx:dx:20.
num_steps = iround(c^2*dt/(2.0*D)); 
probs = cell(M);
probs[1], walker_pos = get_prob_random_walker(x_ends,lambda,num_steps);
for k in 2:M
    probs[k] = add_prob_random_walker(x_ends,walker_pos,lambda,k*num_steps); 
end

final_result = probs[1];
for p in 2:M; 
    final_result += probs[p]; 
end
```

```julia
final_result *= dt*Q0;
```

```julia
plot(x,final_result)
```
