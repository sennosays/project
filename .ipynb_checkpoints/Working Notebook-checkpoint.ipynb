```julia
function parallel_explicit_solver(u0,this_xi::Float64,this_xf::Float64,this_tf::Float64,this_D::Float64,nx::Int = 20)
    @assert(this_xf > this_xi); 
    @assert(this_tf > 0.0); 
    @assert(this_D > 0.0); 

    x = linspace(this_xi,this_xf,nx+2); 
    dx = x[2] - x[1]; 
    @assert(dx == (xf-xi)/(nx+1));
    M = iround(tf*this_D/(0.49*dx^2)); 
    dt = tf/M; 
    lambda = dt*this_D/dx^2; 
    @assert(lambda < 0.5); 
    temp = 1.0-2.0*lambda; 
    old_u = map(u0,x[2:end-1]);
    new_u = Array(Float64,nx); 
    
    for j in 0:M-1
        new_u[1] = lambda*old_u[2] + temp*old_u[1]; 
        @parallel for i in 2:nx-1
            new_u[i] = lambda*(old_u[i+1]+old_u[i-1]) + temp*old_u[i]; 
        end
        new_u[nx] = lambda*old_u[nx-1] + temp*old_u[nx];
        old_u = copy(new_u); 
    end 
    
 
    prepend!(old_u,[0.0]); 
    append!(old_u,[0.0]); 
    
    return old_u, x;
end
```

```julia
@everywhere function parallel_explicit_stepper!(num_steps::Int,this_old_u::DArray,this_lambda::Float64)
    @assert(num_steps > 2); 
    
    u_copy = copy(this_old_u); 
    println("the length of this local part is ", length(this_old_u)); 
    len = length(u_copy); 
    this_new_u = Array(Float64,len); 
    
    temp = 1.0 - 2.0*this_lambda; 
    left_bound = (myid() == workers()[1]); 
    right_bound = (myid() == workers()[end]); 
    for j in 1:num_steps
        if !left_bound && !right_bound
            this_new_u = [this_lambda*(u_copy[i-1] + u_copy[i+1]) + temp*u_copy[i] for i in 1:len]; 
        elseif left_bound
            this_new_u[1] = this_lambda*u_copy[2] + temp*u_copy[1];
            this_new_u[2:end] = [this_lambda*(u_copy[i+1]+u_copy[i-1]) + temp*u_copy[i] for i in 2:len];
        else 
            this_new_u[end] = this_lambda*u_copy[end-1] + temp*u_copy[end]; 
            this_new_u[1:end-1] = [this_lambda*(u_copy[i+1]+u_copy[i-1]) + temp*u_copy[i] for i in 1:len-1];
        end
        u_copy = copy(this_new_u); 
    end 
    localpart(this_old_u) = copy(this_new_u);  
end
```
