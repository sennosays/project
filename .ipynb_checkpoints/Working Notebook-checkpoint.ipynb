```julia
@everywhere function parallel_explicit_stepper!(num_steps::Int,d_u,local_u::Array{Float64,1},this_lambda::Float64)
    this_range = (d_u.indexes[myid()-1])[1];
    len = length(this_range);
    @assert(len == length(local_u)); 
    new_u = Array(Float64,len); 
    
    left_boundary = this_range[1] == 1;
    right_boundary = this_range[end] == length(d_u); 
    middle = !left_boundary & !right_boundary; 
    
    temp = 1.0-2.0*this_lambda; 
    println("hello my range is ",this_range, " and my left_boundary is ",left_boundary); 
    println("my right boundary is ",right_boundary); 
    for j in 1:num_steps
        if middle 
            new_u = [this_lambda*(d_u[i-1] + d_u[i+1]) + temp*d_u[i] for i in this_range];
        elseif left_boundary 
            new_u[1] = this_lambda*d_u[2] + temp*d_u[1]; 
            new_u[2:len] = [this_lambda*(d_u[i-1] + d_u[i+1]) + temp*d_u[i] for i in this_range[2:end]];
        elseif right_boundary
            new_u[len] = this_lambda*d_u[len-1] + temp*d_u[len];
            new_u[1:len-1] = [this_lambda*(d_u[i-1] + d_u[i+1]) + temp*d_u[i] for i in this_range[1:end-1]];
        else
            println("Error with Bounds"); 
            return
        end
        
    end
    local_u;   
end
```

```julia
@everywhere function parallel_explicit_stepper!(num_steps::Int,this_old_u::DArray,this_lambda::Float64)
    @assert(num_steps > 2); 
    
    u_copy = copy(this_old_u); 
    println("the length of this local part is ", length(this_old_u)); 
    len = length(u_copy); 
    this_new_u = Array(Float64,len); 
    
    temp = 1.0 - 2.0*this_lambda; 
    left_bound = (myid() == workers()[1]); 
    right_bound = (myid() == workers()[end]); 
    for j in 1:num_steps
        if !left_bound && !right_bound
            this_new_u = [this_lambda*(u_copy[i-1] + u_copy[i+1]) + temp*u_copy[i] for i in 1:len]; 
        elseif left_bound
            this_new_u[1] = this_lambda*u_copy[2] + temp*u_copy[1];
            this_new_u[2:end] = [this_lambda*(u_copy[i+1]+u_copy[i-1]) + temp*u_copy[i] for i in 2:len];
        else 
            this_new_u[end] = this_lambda*u_copy[end-1] + temp*u_copy[end]; 
            this_new_u[1:end-1] = [this_lambda*(u_copy[i+1]+u_copy[i-1]) + temp*u_copy[i] for i in 1:len-1];
        end
        u_copy = copy(this_new_u); 
    end 
    localpart(this_old_u) = copy(this_new_u);  
end
```

```julia
@everywhere function initialize(x,total_x)
    for i in 1:length(x)
        x[i] = total_x[end];
    end
end
@everywhere function flip_flop(one,two)
    for i in 1:length(one)
        one[i] = myid();
        two[i] = one[i] + 1
    end
    for j in 1:length(one)
        one[j] = two[j];
    end
end     
@everywhere function get_other_piece(a,b)
    
    a[1] = remotecall_fetch(3,getindex,localpart(b),length(localpart(b))); 
end
```
