```julia
function MC_crank_solver(u0,xi::Float64,xf::Float64,tf::Float64,D::Float64,nx::Int = 50,num_walkers::Int = 1000)
    @assert(xf>xi); 
    @assert(tf>0.0); 
    @assert(D > 0.0); 

    x = linspace(xi,xf,nx+2); 
    dx = x[2] - x[1]; 
    @assert(dx == (xf-xi)/(nx+1));
    nt = iceil(tf*D/dx^2); 
    dt = tf/nt; 
    lambda = dt*D/dx^2;
    println("lambda is ",lambda); 
    @assert(lambda < 1.0); 
    
    initial_condition = map(u0,x);
    solution = Array(Float64,nx+2); 
    counts = zeros(Int,nx+2); 
    probs = zeros(Float64,nx+2); 
    probability = (1.0-lambda)/(1.0+lambda); 
    solution[1] = 0.0;
    solution[end] = 0.0;

    for i = 2:nx+1
        boundary_hits = 0; 
        fill!(probs,0.0); 
        fill!(counts,0.0); 
        for n = 1:num_walkers
            position_index = i; 
            time_index = nt; 
            
            while((time_index > 0) && (position_index > 1) && (position_index < nx+2))
                if(rand() < probability)
                    time_index -= 1;
                else 
                    r = rand(); 
                    if r < 0.25 
                        position_index += 1; 
                    elseif r < 0.5 
                        position_index -= 1;
                    elseif r < 0.75
                        position_index += 1; 
                        time_index -= 1; 
                    else
                        position_index -= 1;
                        time_index -= 1; 
                    end
                end
            end
            if time_index > 0
                boundary_hits += 1; 
            else
                counts[position_index] += 1; 
            end
        end
        @assert(sum(counts) + boundary_hits == num_walkers);
        probs = counts./num_walkers; 
        solution[i] = sum(probs.*initial_condition); 
    end

    return solution, x; 
end
```

The equation to solve the advection diffusion equation using a trapezoidal integration method: 
$$\partial_t u = F(x,t,u)$$
$$u_{i,j+1,k} \approx u_{i,j,k} + \frac{dt}{2}\left(F_{i,j+1,k} + F_{i,j,k}\right)$$
where $(i,j,k) \rightarrow $(space, time, energy)

if we make the following definitions
$$\alpha_k \equiv \frac{dt\ D_k}{dx^2}$$

$$\beta_k \equiv \frac{dt\ b_k}{2\ dE}$$

Where $b_k$ is the energy loss rate evaluated at energy point $k$ and $D_k$ gives the diffusion constant for energy point $k$ 

and the approximation 

$$u_{i,j,k} \simeq \frac{1}{4}\left[u_{i,j+1,k+1} + u_{i,j,k+1} + u_{i,j+1,k-1} + u_{i,j,k-1}\right]$$

then we have the following finite difference equation to solve 

\begin{equation}
u_{i,j+1,k} = \frac{\alpha_k}{2(1+\alpha_k)}\left[u_{i+1,j+1,k} + u_{i+1,j,k} + u_{i-1,j+1,k} + u_{i-1,j,k} \right] + 
\frac{1/4-\alpha_k/4+\beta_{k+1}/2)}{1+\alpha_k}\left[u_{i,j+1,k+1} + u_{i,j,k+1}\right] + \frac{1/4-\alpha_k/4-\beta_{k-1}/2)}{1+\alpha_k}\left[u_{i,j+1,k-1} + u_{i,j,k-1}\right]
\end{equation}

notice that we have the condition 

$$\frac{1}{4} - \frac{\alpha_k}{4} - \beta_{k-1}/2 > 0$$



after normalization the probability of picking a direction from the first term is $1/4 \times$

$$ \frac{\alpha_k}{1 + \alpha_k + \beta_{k+1} -\beta_{k-1}}$$

and the other two terms have a probability of $1/2 \times$ 

$$ \frac{\frac{1}{2} - \frac{\alpha_k}{2} + \beta_{k-1}}{1 + \alpha_k + \beta_{k+1} -\beta_{k-1}}$$

and

$$ \frac{\frac{1}{2} - \frac{\alpha_k}{2} - \beta_{k-1}}{1 + \alpha_k + \beta_{k+1} -\beta_{k-1}}$$


respectively 

```julia
xi = 0.0; 
xf = 50.0; 
nx = 100; 
x = linspace(xi,xf,nx);
dx = x[2]-x[1];

Ei = 0.0; 
Ef = 20.0;
nE = 70;

spec_E_min = 5.0;
spec_E_max = 17.;
E = linspace(Ei,Ef,nE); 
dE = E[2] - E[1];

alpha = map(calc_diffusion_constant,E);
beta = map(calc_energy_loss_rate,E);

tf = 0.1; 
nt = iceil(tf*maximum(alpha./dx^2 + beta/dE)); 
dt = tf/nt;

alpha *= (dt/dx^2);  
beta *= (dt*0.5/dE); 
beta = - beta;

@assert(minimum(0.5.-0.5*alpha - beta) > 0)

initial_condition = [energy_spectrum(a,spec_E_min,spec_E_max)*u0(b,0.5*(xf-xi),1.0) for b=x, a=E];
solution = Array(Float64,nx,nE); 
counts = Array(Int,nx,nE); 
probs = zeros(Float64,nx,nE); 
```

```julia
probabilities = Array(Float64,nE,3); 
normalization = 1.0+alpha[1] + beta[2]
probabilities[1,:] = [alpha[1]/normalization, (0.5 - 0.5* alpha[1] + beta[2])/normalization, (0.5 - 0.5*alpha[1])/normalization]; 

for k in 2:nE - 1
    normalization = 1 + alpha[k] + beta[k+1] - beta[k-1]; 
    probabilities[k,:] = [alpha[k]/normalization, (0.5 - 0.5* alpha[k] + beta[k+1])/normalization, (0.5 - 0.5*alpha[k] - beta[k-1])/normalization];
end

normalization = 1.0 + alpha[nE] - beta[nE-1]; 
probabilities[nE,:] = [alpha[nE]/normalization, (0.5 - 0.5* alpha[nE])/normalization, (0.5 - 0.5*alpha[nE] - beta[nE-1])/normalization];
probabilities *= 0.5;

for k in 1:nE 
    @assert(abs((4.0*probabilities[k,1] + 2.0*probabilities[k,2] + 2.0*probabilities[k,3]) - 1.0) < 0.0001)
end
```

```julia
num_walkers = 750;
for i = 2:nx-1
    if( i%(0.1*nx) == 0)
        println(100*i/nx,"% completed"); 
    end
    for k = 2:nE-1
        boundary_hits = 0; 
        fill!(probs,0.0); 
        fill!(counts,0.0); 
        for n = 1:num_walkers
            position_index = i; 
            energy_index = k; 
            time_index = nt; 
            
            while((time_index > 0) && (position_index > 1) && (position_index < nx) && (energy_index > 1) && (energy_index < nE))
                r1 = rand() 
                if(r1 < 4.0*probabilities[k,1])
                    r2 = rand(); 
                    if r2 < 0.25 
                        position_index += 1; 
                    elseif r2 < 0.5 
                        position_index += 1;
                        time_index -= 1;
                    elseif r2 < 0.75
                        position_index -= 1; 
                    else
                        position_index -= 1;
                        time_index -= 1; 
                    end
                elseif(r1 < (4.0*probabilities[k,1] + 2.0*probabilities[k,2]))
                    r2 = rand();
                    energy_index += 1; 
                    if(r2 < 0.5)
                        time_index -= 1; 
                    end
                else
                    r2 = rand(); 
                    energy_index -= 1; 
                    if(r2 < 0.5); 
                        time_index -= 1; 
                    end
                end
            end
            if time_index > 0
                boundary_hits += 1; 
            else
                counts[position_index,energy_index] += 1; 
            end
        end 
        @assert(sum(counts) + boundary_hits == num_walkers);
        #println("the total number of counts was ",sum(counts)," which is ",100*sum(counts)/num_walkers,"% of all the walkers"); 
        probs = counts./num_walkers; 
        solution[i,k] = sum(probs.*initial_condition); 
    end
end
soultion[1,:] = 0.0; 
solution[end,:] = 0.0; 
soltuion[:,1] = 0.0; 
solution[:,end] = 0.0; 
println(100.0,"% completed"); 
```
